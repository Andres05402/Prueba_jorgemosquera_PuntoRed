Configuración de variables del notebook

```python
catalog_name = "jorge_catalog"
schema_name = "transacciones"
volume_name = "volumen_prueba"
file_path = f"/Volumes/{catalog_name}/{schema_name}/{volume_name}/transactions_50k.jsonl"
bronze_path = f"/Volumes/{catalog_name}/{schema_name}/{volume_name}/bronze/transactions"

```

lee jsonl

```python
df_bronze = spark.read.json(file_path, multiLine=False)
```

Guarda jsonl

```python
#df_bronze.write.format("delta").mode("overwrite").save(bronze_path)
df_bronze.write.format("delta").mode("overwrite").saveAsTable("jorge_catalog.transacciones.bronze")
```

Validacion de cargue

```python
# Lectura desde Bronze para validar
df_check = spark.read.format("delta").load(bronze_path)
print(f"Total registros en Bronze: {df_check.count()}")
```
